# Latest insights generated

```python
from tiny_data_warehouse import DataWarehouse

tdw = DataWarehouse()


df = tdw.event('sota_paper_insight').sort_values(by='tdw_timestamp', ascending=False)
DATE_FROM='2024-07-01'
df = df[df['tdw_timestamp'] > DATE_FROM]
df.to_dict()

for i, row in df.iterrows():
    print(row['insights'][0:50])
    print(row['tdw_timestamp'])
    print(row['abstract_url'])
    print("PDF: ", row['pdf_path'])
    print("")
```
