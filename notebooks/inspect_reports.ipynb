```python
from tiny_data_warehouse import DataWarehouse
tdw = DataWarehouse()
df = tdw.event('state_of_the_art_summary').sort_values(by='tdw_timestamp', ascending=False)
df
```

```python
df['papers_analysed_count'] = df['papers_analysed'].apply(lambda x: len(str(x).split('\n')))
```

```python
from state_of_the_art.paper.url_extractor import PapersUrlsExtractor

extractor = PapersUrlsExtractor()

df['output_papers'] = df['summary'].apply(lambda x: extractor.extract_urls(str(x)))
```

```python
df['output_papers_top_ten'] = df['output_papers'].apply(lambda x: x[:10])

df['papers_analysed_top_ten'] = df['papers_analysed'].apply(lambda x: str(x).split("\n")[:10])

df['overlapping_papers'] = df.apply(lambda x: len(set(x['output_papers_top_ten']).intersection(set(x['papers_analysed_top_ten']))), axis=1)

df.head(100)
```
